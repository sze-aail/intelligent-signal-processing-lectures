## Visualization
This is an example of a visualization of a random matrix.

```python
X = np.random.rand(256, 256)
print(X)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(X, cmap='hot', interpolation='nearest')
```

```python
Y = np.random.rand(256, 256)
plt.imshow(X @ Y, cmap='hot', interpolation='nearest')
```

### Plotting

Let's plot the different functions, linear, quadratic, cubic, and exponential.

```python
t = np.linspace(0, 10, 100)
```

```python
plt.plot(t, t, label='linear')
plt.plot(t, t**2, label='quadratic')
plt.plot(t, t**3, label='cubic')
plt.plot(t, 2**t, label='exponential')
plt.legend()
plt.ylim(0, 30)
plt.xlabel('t')
plt.ylabel('f(t)')
# Origo point
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid()
plt.show()
```
